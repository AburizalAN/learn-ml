```python
total = airbnb_df.isnull().sum().sort_values(ascending=False)
count = airbnb_df.isnull().count().sort_values(ascending=False)
percentage = (total / count * 100).sort_values(ascending=False)

missing_data = pd.concat([total, percentage], axis=1, keys=["Total", "percentage"])

missing_data
```
