```python
import pandas as pd

url='https://storage.googleapis.com/rg-ai-bootcamp/visualization/airbnb-data.csv'
airbnb_df = pd.read_csv(url)

airbnb_df
```

```python
avg_prices_per_year = airbnb_df.groupby('construction_year')['price'].mean()
years = avg_prices_per_year.index
prices = avg_prices_per_year.values

print(years)
print(prices)
```
